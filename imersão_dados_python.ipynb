```python
#importando a biblioteca numpy para manipulaçao de numeros
import numpy as np

#Em resumo, este código cria uma pequena "tabela" (um DataFrame)
#chamada df_salarios com os nomes e salários de quatro pessoas, onde os salários
#de duas delas estão marcados como ausentes.
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 10000]
})

#cria uma nova coluna chamada salario_media e preenche os salários que estão
#faltando (NaN) com a média arredondada de todos os outros salários que existem.
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#exatamente o mesmo comando da linha acima, diferença é que agora estamos
#pegando a mediana dos valores.
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

#chama a coluna nova que criamos
df_salarios
```

```python
#np.nan significa "NumPy Not a Number"
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#ffill preenche valores nulos com o valor anterior da mesma tabela
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas
```

```python
#np.nan significa "NumPy Not a Number"
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

#bfill preenche valores nulos com o valor posterior da mesma tabela
df_temperaturas['preenchido_bffill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
```

```python
#para criar dicionarios se usam {}
#para criar listas se usam []
#para criar tuplas se usam ()
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['Sao Paulo', np.nan, 'Curitiba', np.nan, 'Belem']
})

#preenche os valores nulos com a string que informamos entre parenteses
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Nao informado')

df_cidades
```

```python
#cria um novo DataFrame chamado df_limpo contendo apenas as linhas do DataFrame
#original (df) que não possuem nenhum valor ausente (NaN) em nenhuma de suas
#colunas. É a forma mais rápida de remover todas as linhas "problemáticas" que
#contêm dados faltantes.
df_limpo = df.dropna()
```

```python
#forma padrão de verificar se a sua limpeza de dados foi bem-sucedida ou para
#diagnosticar quais colunas ainda precisam de tratamento.
df_limpo.isnull().sum()
```

```python
#calcula a frequência de cada categoria na coluna 'senioridade' e, em seguida,
#gera um gráfico de barras para visualizar essa distribuição.
df_limpo['senioridade'].value_counts().plot(kind='bar', title='Distribuição Senioridade')
```

```python
#seaborn é uma biblioteca de visualizaçao de dados construida em cima de outra
#biblioteca chamada Matplotlib. Seaborn torna a criação de gráficos estatísticos
#atraentes e informativos muito mais fácil.
import seaborn as sns
```

```python
#cria um gráfico de barras que mostra a relação entre uma variável categórica
#('senioridade') e uma variável numérica ('usd').
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')
```

```python
#Matplotlib é como a "mãe" de todas as bibliotecas de visualização em Python.
#É a base sobre a qual muitas outras, incluindo o Seaborn são construídas.
#pyplot é um submódulo (uma parte específica) da matplotlib. É a coleção de
#funções que torna a criação de gráficos muito mais simples e direta.
import matplotlib.pyplot as plt
```

```python
#cria uma figura definindo o tamanho com figsize
plt.figure(figsize=(8,5))

#funça para criar um grafico de barras
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd')

#define um titulo para a tabela
plt.title('Salário médio por Senioridade')

#define um nome para o eixo x = horizontal
plt.xlabel('Senioridade')

#define um nome para o eixo y = vertical
plt.ylabel('Salario médio anual (USD)')

#exibe o resultado da consulta
plt.show()
```

```python
#calcula o salário médio (usd) para cada nível de senioridade e, em seguida,
#exibe esses resultados ordenados do maior salário médio para o menor.
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = False)
```

```python
#cria uma variavel e indexa os niveis de senioridade na variavel
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = False).index

#chamar e mostrar a variavel
ordem
```

```python
#cria uma figura definindo o tamanho com figsize
plt.figure(figsize=(8,5))

#funça para criar um grafico de barras chamando a variavel
sns.barplot(data = df_limpo, x = 'senioridade', y = 'usd', order = ordem)

#define um titulo para a tabela
plt.title('Salário médio por Senioridade')

#define um nome para o eixo x = horizontal
plt.xlabel('Senioridade')

#define um nome para o eixo y = vertical
plt.ylabel('Salario médio anual (USD)')

#exibe o resultado da consulta
plt.show()

#perceba que agora as barras estao de forma ascendente e de melhor visualizaçao.
```

```python
#cria uma figura definindo o tamanho com figsize
plt.figure(figsize = (10,5))

#cria um histograma da coluna de salários (usd), dividindo a faixa de salários
#em 50 "caixas" (bins) para mostrar a frequência de salários em cada faixa,
#e sobrepõe uma linha suave (KDE) para ajudar a visualizar a forma geral
#da distribuição.
sns.histplot(df_limpo['usd'], bins = 45, kde = True)

#define um titulo para a tabela
plt.title('Distribuição dos salários anuais')

#define um nome para o eixo x = horizontal
plt.xlabel('Salário em USD')

#define um nome para o eixo y = vertical
plt.ylabel('Frequencia')

#exibe o resultado da consulta
plt.show()
```

```python
#cria uma figura definindo o tamanho com figsize
plt.figure(figsize = (8,5))

#cria um boxplot (gráfico de caixa) para a coluna de salários (usd), que
#resume a distribuição dos dados, mostrando a mediana, os quartis e,
#crucialmente, identificando valores atípicos (outliers).
sns.boxplot(x = df_limpo['usd'])

#define um titulo para a tabela
plt.title('Boxplot Salário')

#define um nome para o eixo x = horizontal
plt.xlabel('Salário em USD')

#exibe o resultado da consulta
plt.show()
```

```python
#cria uma lista
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

#cria uma figura definindo o tamanho com figsize
plt.figure(figsize = (8,5))

#cria e exibe múltiplos boxplots lado a lado, um para cada nível de senioridade,
#mostrando a distribuição dos salários (usd) para cada um desses níveis.
#Crucialmente, ele ordena os boxplots na tela de acordo com a
#lista ordem_senioridade.
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade)

#define um titulo para a tabela
plt.title('Boxplot da distribuiçao por senioridade')

#define um nome para o eixo x = horizontal
plt.xlabel('Salário em USD')

#exibe o resultado da consulta
plt.show()
```

```python
#cria uma lista
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

#cria uma figura definindo o tamanho com figsize
plt.figure(figsize = (8,5))

#cria e exibe múltiplos boxplots lado a lado, um para cada nível de senioridade,
#mostrando a distribuição dos salários (usd) para cada um desses níveis.
#Crucialmente, ele ordena os boxplots na tela de acordo com a
#lista ordem_senioridade adicionando cor em cada uma delas.
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade, palette = 'Set2', hue = 'senioridade')

#define um titulo para a tabela
plt.title('Boxplot da distribuiçao por senioridade')

#define um nome para o eixo x = horizontal
plt.xlabel('Salário em USD')

#exibe o resultado da consulta
plt.show()
```

```python
#a biblioteca plotly gera gráficos dinâmicos baseados em tecnologias web como
#HTML e JavaScript. Isso significa que você, ou quem estiver vendo seu gráfico,
#pode interagir com ele.
#express é o submódulo que você quase sempre vai querer usar. Ele foi criado
#para ser conciso, claro e fácil de aprender.
import plotly.express as px
```

```python
#este trecho de código foi gerado por IA
#prompt: crie um grafico de media salarial por senioridade em barras usando plotly
# Calcula a média salarial por senioridade
media_salario_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

# Cria a lista de ordem desejada para as barras
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

# Cria o gráfico de barras interativo com Plotly Express
fig = px.bar(media_salario_senioridade,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade (USD)',
             labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual (USD)'},
             category_orders={'senioridade': ordem_senioridade}) # Define a ordem das categorias

# Exibe o gráfico
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()

remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

# Cria o gráfico de pizza interativo com Plotly Express
fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho'
             )

# Exibe o gráfico
fig.show()
```

```python
#utiliza o pip, o gerenciador de pacotes padrão do Python, para baixar e
#instalar a biblioteca pycountry, que serve como um banco de dados de
#padrões ISO (como países, estados, moedas e idiomas) diretamente acessível
#dentro do seu código Python.
pip install pycountry
```

```python
import pycountry

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
